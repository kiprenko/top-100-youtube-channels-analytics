## importing libs and dataset

```python
import pandas as pd

df = pd.read_csv('top-100-youtube-channels-dataset.csv')
# removing extra whitespaces in column names
df = df.rename(columns=lambda x: x.strip())
# replacing whitespaces with underscores
df = df.rename(columns=lambda x: x.replace(' ', '_'))
df.head()
```

### Sorting by sibscribers just in case

```python
df.sort_values(by=['video_views']);
df.head()
```

### Adding a new column called "views_rate"
**Description:** the column is calculated in the next way: *video_views / (video_count * subscribers)*

```python
df = df.assign(views_rate=lambda x: x.video_views / (x.video_count * x.subscribers))
df.head()
```

### Sorting value by the "views_rate" column

```python
df = df.sort_values(by='views_rate', ascending=False);
top_20_in_general = df.head(20)
top_20_in_general
```

### Saving TOP-20 best channel in general to csv file

```python
top_20_in_general.to_csv("top-20-in-general.csv")
```
